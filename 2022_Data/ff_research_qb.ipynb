### Josh Allen

```python
Image(url="ja.jpg")
```

In 2022, Josh Allen was the highest drafted QB. It made sense, since he was the QB1 (1st ranked QB in terms of fantasy points scored) for two seasons straight. Football analysts both inside and outside the fantasy world debated whether he might be the NFL's best quarterback over Patrick Mahomes. Since 2020, Josh Allen's passing stats were elite, but behind Mahomes. However, Allen was twice as likely to run the ball, and more effective than Mahomes when doing so. His running style was as exciting as it was unconventional, as he elected to run through defenders as opposed to around them, sometimes surprising people by jumping over them completely. Big and athletic, this becames a staple of his game. It all culminated in an exciting playoff game in the 2021 season against Mahomes' Chiefs where both quarterbacks put on a show. When Buffalo lost in overtime without Josh Allen even touching the ball, it caused the NFL to change the overtime rules for future playoffs.

This year, Josh Allen was the odds favorite to win MVP. The Bills were the favorite to win the superbowl. Buffalo made their team stronger by boosting their already great defense in free agency, and drafting a running back to help the offense. They also expected their promising young receiver (Gabriel Davis) to take the next step in his development. On the other hand, Patrick Mahomes lost his best receiver (Tyreek Hill) to Miami in the offseason. Operating an offense where the best pass catcher is an aging TE in his 9th year would be a challenge for any quarterback, but it was one Kansas City was comfortable with.

So how did Josh Allen perform in 2022? As expected? Better than expected? Worse then expected?

In short, he played well. As the QB2 for this season (QB3 on a per-game basis), I would say he performed as expected. Its possible you wanted the top-ranked quarterback in the fantasy draft to be an automatic QB1. However, whenever I draft a top QB (or TE, RB, WR), I expect stability. Things happen in any given season. Josh Allen was top 3 in total points and points per game. The QB1 in total points won MVP, and the QB1 in points per game finished second in MVP voting, just ahead of Josh Allen.

So while he did play well, it wasn't perfect. Let's see what worked against him, and if it is cause for concern for future seasons.

```python
df_ja = df2[df2['TEAM']=='BUF']
df_ja = df_ja[['NAME','TEAM','FPTS','ATT','COMP','YDS','TD','INT','300+','RUSH','RUSHYDS','RUSHTD','WEEK']]
df_ja.describe()
```

```python
df_ja
```

These charts mainly paint Josh Allen in a favorable light. The second one shows his stats for each week. I tend to expect my QB (superstar or not) to score 20 fantasy points (FPTS) for me to consider it a good week. Allen crossed that threshold consistently, missing the mark only 3 times in 2022. He only had one 'bad' performance in week 11, scoring only 12.6 FPTS.

However, I do remember analysts pointing out Josh Allen not playing up to his standard halfway through the season. If you split the stats, you may find some interesting discrepancies. Josh Allen's stats look a bit different before and after Buffalo's week 7 bye week

```python
df_ja_prebye = df_ja.iloc[[0,1,2,3,4,5]]
df_ja_prebye
```

```python
df_ja_prebye.describe()
```

```python
df_ja_postbye = df_ja.iloc[[6,7,8,9,10,11,12,13,14]]
df_ja_postbye
```

```python
df_ja_postbye.describe()
```

I will preface this by saying an average of 23.32 FPTS per game is still really good (for fantasy football). However since Allen was drafted so high, we're allowed to nitpick a little. The difference in Allen's performance was over 6 FPTS per game. But why? His passing stats took a hit, but his rushing stats were a bit better (not enough to make up the FPTS difference). Also, there's a good chance the coaching staff would preferred if he ran a little less as his aggressive running style subjugated him to big hits from opposing defenses.

Two notable stats that show his drop in performance are actually not listed in our stat table, but use the numbers above, are completion percentage (COMP/ATT) and TD-to-INT ratio (TD/INT).

Josh Allen's completion percentage dropped from 67.3 to 60.36 after Buffalo's bye week, which may not seem like much, until you check out the completion percentage of other QBs in the league (https://sports.yahoo.com/nfl/stats/individual/?sortStatId=COMPLETION_PERCENTAGE&selectedTable=0). If Josh Allen maintained 67.3 for the season, he would've ranked 5. 60.36 ranks 31st. Altogether, his rate of 63.3 was in the bottom half of the league. He also threw 17 TDs against 4 INTs before the bye. Thats more than 4 TDs for every INT. After the bye, that rate dropped to 1.8.

So Josh Allen was far less efficient passing the ball after the teams bye week. Was there a notable change in the offense? Well, Josh Allen did suffer an injury in his shoulder. While he did play through it, it wouldn't be surprising if it affected his passing a bit.

However that injury happened towards the end of Buffalo's week 9 game. If we were to divide the season after week 9, things look very different.

```python
df_ja_preinj = df_ja.iloc[[0,1,2,3,4,5,6,7]]
df_ja_preinj.describe()
```

```python
df_ja_postinj = df_ja.iloc[[8,9,10,11,12,13,14]]
df_ja_postinj.describe()
```

The difference pre injury and post injury is that Josh Allen threw less after hurting his shoulder. His completion percentage and TD-to-INT ratio look very similar. While there was an argument that this injury affected his accuracy, there was also an argument that Josh Allen will be too aggressive on occasion. That same aggressiveness is the reason Allen is as great as he is, so you have to take the bad with the good.

So we have two charts (pre-bye, post-bye) suggesting an efficient super Josh Allen is possible. We also have two charts (pre-injury, post-injury) that suggest a less than efficient but healthier Josh Allen may have been enough to grab QB1 for the third straight year. We need a tie breaker.

https://sports.yahoo.com/nfl/players/30977/

After this season, Josh Allen has played 5 seasons in the NFL. The first 2 were learning experiences. His third year (2020) is when he "broke out" and became a superstar. It was his best year throwing the football, with career highs in most categories, including efficiency. 69.2 completion rating was fourth best that season (would've been second best this season). He decidedly ran the ball a lot less. In his fourth season, his passing stats were down, probably due to defenses taking him a lot more seriously, but he supplemented that with career highs running the ball. This season (his fifth) was statistically similar to last season, in some ways even better. This is with a shoulder injury slowing him down.

So efficient Josh Allen is more of a fantasy. But aggressive (and healthy) Josh Allen is really good too. I don't see why he can't claim the QB1 title next year. Personally, I don't have a (recent) history of drafting QBs early, Josh Allen is a great choice for someone who doesn't want to have to guess who the next great QB might be. However, if he drops in the rankings and is available in a reasonable round, I'd have no trouble picking him (depending on Buffalo's offseason moves).

Buffalo has tried for years to build a good running game to complement. They'll probably try again this offseason. Perhaps a really good running back might take away from Allen, but probably not too much. They could also use another receiver to complement Steon Diggs. Diggs' arrival triggered Allen's break out year in 2020. Another high tier receiver can take this team to the next level. Gabe Davis was thought to be that guy, but this year proved otherwise.

### Lamar Jackson

```python
Image(url="lj.jpg", width=948, height=632)
```

Drafting Lamar Jackson as a QB2 means chasing his ceiling. His ceiling was his 2019 MVP season where he threw for 3127 yards with 36 TDs against only 6 Int at a 66.1 completion rate, in addition to running for 1206 yards and 7 TDs. While those numbers draw Michael Vick comparisons, almost all of them beat Vick's career highs. All of this happened in only his second year. In fantasy, he was the QB1 on the year, scoring nearly 75 points more than the next guy. Fantasy tends to favor rushing QBs, but Lamar Jackson has taken great strides to improve his already underrated passing. He's aware of the threat of him running, and uses that to manipulate the defense. He would do well with a top receiver, which Baltimore believed they had with the emerging Rashod Bateman. They were so confident, they traded away Marquise Brown, a receiver Lamar Jackson had success with.

So how did 2022 turn out for Jackson? Well, not good. He was QB13 on the year (QB6 for per-game)

```python
df_lj = df2[df2['TEAM']=='BAL']
df_lj = df_lj[['NAME','TEAM','FPTS','ATT','COMP','YDS','TD','INT','300+','RUSH','RUSHYDS','RUSHTD','WEEK']]
df_lj.describe()
```

```python
df_lj
```

So the first three weeks were more than promising. Jackson scored at a perhaps unsustainable rate. But it all fell apart afterward. What happened after week three? The key to a great fantasy season for Lamar was always going to start with Bateman, but Bateman only plays in six games this season. He starts missing games week 5, and doesn't play much week 4 due to a foot injury.

It should be noted while Lamar struggled in fantasy, he did win 8 of his 12 games.

A healthy Jackson and Bateman could make for a good 2023 season in Baltimore. They've also hired a new offensive coordinator whose been known to be more aggressive throwing downfield. This could help elevate Jackson's passing game. Baltimore would also help themselves out by adding more receiver depth, so Jackson won't be scraping by to get a win.
